We now have an implementation to compute the Riemann zeta function and would like to study its properties. 
To get a rough idea of what to expect it is sometimes useful to make some plots. (If you dont have a sucessful implementation to compute zeta you can use the builtin function at this stage if you like. 

One option would be to use the 3d plot:

```python
# By sucessive plots in various ranges we find that the first zero seems to be around 1/2 + 14*I 
# 
var('x','y')
plot3d(lambda x,y: abs(zeta(CC(x,y))),(x,0,1),(y,10,15))
```

Note that a plot including the point (x,y)=(0,1) doesn't show due to the pole at x=1.

One of the problems that usually arises in scientific writing is how to include graphs and plots.
In SageMath there is a very useful `LaTeX` export function using PGF which unfortunately only works for simple plots.

```python
p1 = plot(sin,0,10);p1
```

```python
latex(p1)
```

```python
p2=plot3d(lambda x,y: abs(zeta(CC(x,y))),(x,0,1),(y,10,15))
# There are different viewer and some might not work in your browser
# e.g. `threejs` does not work on my laptop... 
p2.show(viewer='canvas3d')
```

```python
latex(p2)
```

As an alternative here we can save the figure as png:

```python
p2.save_image('test.png')
```

Note that 2-d plots can be saved as pdf but 3d plots not

```python
p1.save_image('test1.pdf')
```
